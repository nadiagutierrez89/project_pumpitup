# Qué hace este notebook???

En este notebook se entrena una nueva red neuronal desde cero, usando un dataset de spectrogramas y datos de pasos asociados.

Luego se prueba la red con algunas imágenes de ejemplo.

```python
from ai import PumpIA

pia = PumpIA()
pia.load_dataset(
    dataset_json_path="./resultado_con_cant_notas.json", 
    train_sgrams_root_path="./sgrams",
    sample_size=14000,
)

pia.show_sgrams(sample=3)
```

```python
pia.build_and_train_neural_network(train_epochs=5)
pia.save("./ia_entrenada.pkl")
```

```python
# solo obtener una predicción
pia.predict("./sgrams/PE62_52.png")
```

```python
# obtener una predicción pero mirar un poco más en detalle la imagen, la predicción cruda interna, etc
pia.show_and_predict("./sgrams/PE62_63.png")
```

```python
from pathlib import Path
from random import sample

test_sgrams = sample(list(Path("./sgrams/").glob("*.png")), 10)

for sgram in test_sgrams:
    pia.show_and_predict(sgram)
```
