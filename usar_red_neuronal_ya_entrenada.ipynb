# Qué hace este notebook???

En este notebook se muestra cómo se puede usar una red neuronal ya entrenada. Para ello hace falta tener el archivo `ia_entrenada.pkl`, que te lo puede pasar alguien que ya haya entrenado la red neuronal, o podés generarlo vos usando el otro notebook.

```python
from ai import PumpIA

pia = PumpIA.load("./ia_entrenada.pkl")
```

```python
# solo obtener una predicción
pia.show_and_predict("./output_taylor/taylor_120.png")
```

```python
from pathlib import Path

full_file = []
for sgram_path in Path("./output_taylor/").glob("*.png"):
    full_file.append(pia.predict(sgram_path))
```

```python
for line in full_file:
    print(line[:5], line[5:10], line[10:15], line[15:], sep="\n")
    print(",")
```
